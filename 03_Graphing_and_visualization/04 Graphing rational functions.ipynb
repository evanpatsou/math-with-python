# Graphing rational functions

- What a rational function is
- More on plotting in Python.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = range(-3,4)
y = np.zeros(len(x))


for i in range(0,len(x)):
    y[i] = 2 - x[i]**2
    
plt.plot(x,y,'s-')
plt.xlabel('x'), plt.ylabel('y')
plt.show()
```

```python
# what if you want more spacing?

x = np.linspace(-3,4,100)
y1 = 2 + abs(x)**(1/2)
y2 = 2 + abs(x)**(1/3)
    
plt.plot(x,y1,'m-')
plt.plot(x,y2,'k-')
plt.show()
```

### Exercise

```python
# what if you want more spacing?

x = np.linspace(-4,4,100)
y = [x**i for i in range(-1,4)]

for index in range(0, len(y)):
    plt.plot(x,y[index], label='$y=x^{%s}$'%(index-1),linewidth=4)

plt.axis([-4,4, -20, 20])
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
