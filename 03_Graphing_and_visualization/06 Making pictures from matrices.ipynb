# Making pictures from matrices

- How images are represented as matixes in Python.
- How to visualize matrices using imshow.

```python
# import required packages at the top of the script!
import sympy as sym
import numpy as np
from IPython.display import display, Math
import matplotlib.pyplot as plt
```

```python
# create a matrix
A = [ [1,2],[1,4] ]

# show it (yikes! many functions!)
display(Math(sym.latex(sym.sympify(np.array(A)))))

# now image it
plt.imshow(A)

plt.xticks([0,1])
plt.yticks([.85,1])

plt.show()
```

```python
A = np.zeros((10,14))

print( np.shape(A) )

for i in range(0,np.shape(A)[0]):
    for j in range(0,np.shape(A)[1]):
        A[i,j] = 3*i-4*j

print(A)
plt.imshow(A)
plt.plot([0,3],[8,2],'r',linewidth=4)
plt.set_cmap('Purples')

for i in range(0,np.shape(A)[0]):
    for j in range(0,np.shape(A)[1]):
        plt.text(j,i,int(A[i,j]),horizontalalignment='center',verticalalignment='center')


plt.show()
```

```python
# Create a 10x10 checkerboard
C = np.zeros((10,10))

for i in range(0,np.shape(C)[0]):
    for j in range(0,np.shape(C)[1]):
        C[i,j] = (-1)**(i+j)

plt.imshow(C)
plt.set_cmap('gray')
plt.tick_params(labelleft=False,labelbottom=False)
plt.show()
```
