# Plotting coordinates on a plane

- Get an introduction to graphing in Python.
- Know how to plot single and multiple coordinates.
- Know how to make plots look nicer than the default.

```python
import sympy as sym
import numpy as np
from IPython.display import display, Math
import matplotlib.pyplot as plt
```

```python
x = 3
y = 5

# basic plotting a red dot
plt.plot(x,y,'ro')

# set axis limits
plt.axis('square') # order matters
plt.axis([-6,6,-6,6])
plt.grid()

plt.show()
```

```python
# a set of coordinates

x = [-4,2,5,6,2,-5]
y = [5,2,10,-5,4,0]

for i in range(0, len(x)):
    plt.plot(x[i],y[i], 'o', label='point %s' %(i+1))
    
plt.axis('square')
plt.grid()
plt.legend()

plt.show()
```

```python
# getting information from axes

plt.plot(4,2,'rs')

# get an object for the current axis
axis = plt.gca()
ylim = axis.get_ylim()
print(ylim)

# now change only the upper y-axis limit
axis.set_ylim([ ylim[0],6 ])

plt.xlabel('X axis')
plt.ylabel('F(x)')

plt.show()
```

```python
# Intialize the x,y axis
x = sym.symbols('x')
y = x**2 - 3*x

for i in range(-10,12):
    plt.plot(i, y.subs({x: i}), 'o')

plt.xlabel('x')
plt.ylabel('$f(x) = %s$' %sym.latex(y))
plt.show()
```
