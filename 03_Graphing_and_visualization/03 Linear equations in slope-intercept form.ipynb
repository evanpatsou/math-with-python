# Linear equations in slope-intercept form

- How to describe a line using slope-intercept form.
- More on plotting lines in Python.

```python
import sympy as sym
import numpy as np
from IPython.display import display, Math
import matplotlib.pyplot as plt
```

```python
# The goal: y = mx + b
x = [-5, 5]
m = 2
b = 1

# initialize the y values
y = [314, 4542]

# calculate the yi
for index, xi in enumerate(x):
    y[index] = m*xi + b
    

plt.plot(x,y,label='y=%sx+%s' %(m,b))

plt.axis('square')
plt.grid()
plt.xlim(x)
plt.ylim(x)

axis = plt.gca()
plt.plot(axis.get_xlim(), [0,0], 'k--')
plt.plot([0,0], axis.get_ylim(), 'k--')

plt.legend()

plt.show()
```

```python
# converting x into a numpy array
y = m*np.array(x) + b


plt.plot(x,y,label='y=%sx+%s' %(m,b))

plt.axis('square')
plt.grid()
plt.xlim(x)
plt.ylim(x)

axis = plt.gca()
plt.plot(axis.get_xlim(), [0,0], 'k--')
plt.plot([0,0], axis.get_ylim(), 'k--')

plt.legend()

plt.show()
```

### Exercises

```python
x = [-5, 5]
m = [0.7, -1.25]
b = [-2, 0.75]

# converting x into a numpy array
y = m*np.array(x) + b


for index in range(0,2):
    y = m[index]*np.array(x) + b[index]
    plt.plot(x,y,label='y=%sx+%s' %(m[index],b[index]))

plt.axis('square')
plt.grid()
plt.xlim(x)
plt.ylim(x)

axis = plt.gca()
plt.plot(axis.get_xlim(), [0,0], 'k--')
plt.plot([0,0], axis.get_ylim(), 'k--')

plt.legend()

plt.show()
```
